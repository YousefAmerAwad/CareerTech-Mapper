## 3. One-Hot Encoding

### Objective

- Convert multi-label columns (`skills`, `DevType`) into a machine-learning-ready format using one-hot encoding.

- This step prepares the dataset for modeling

```python
# Function to extract all unique values from a multi-label column

def get_unique_values(column):
    unique_values = set()
    for row in column:
        for value in row:
            unique_values.add(value)
    unique_values_list = list(unique_values)
    unique_values_list.sort(key=str.lower)
    return unique_values_list

```

```python
# Function to perform one-hot encoding on a multi-label column
# Creates binary columns for each unique value (1 if present, else 0).

# Returns: a list of generated column names.
# which can later be used to separate feature columns from target columns.
def one_hot_encode(column):
    unique_values  = get_unique_values(column)

    # initialize all new columns with 0
    df[unique_values] = 0
    
    # fill columns with 1 if the value exists in the row, else 0
    for col in unique_values:
        df[col] = column.apply(lambda x:1 if col in x else 0)

    # return column names (useful later to separate features and target)
    return unique_values
```

```python
# Apply one-hot encoding on the 'skills' column
# and returns the list of generated feature column names
feature_columns  = one_hot_encode(df['skills'])

# Display feature columns
df[feature_columns].head()
```

```python
# Apply one-hot encoding on the 'DevType' column
# and returns the list of generated Target column names
target_columns  = one_hot_encode(df['DevType'])

# Display target columns
df[target_columns]
```

```python
# Remove original multi-label columns after encoding
# They are no longer needed since we now have one-hot encoded features
df.drop(columns=['skills', 'DevType'], inplace=True)

# Verify the result
df.head()
```

<br>

## 4. Feature Scaling

### Objective

Standardize feature values to ensure they are on a comparable scale.<br>

Although one-hot encoded features are binary (0/1), scaling is applied here to maintain consistency across all features and prepare the dataset for models sensitive to feature magnitude.

We will use Z-score normalization:
- Mean = 0
- Standard deviation = 1

```python
# Apply Z-score standardization on feature columns
df[feature_columns]=df[feature_columns].apply(zscore)
df.head()
```

<br>

## 5. Saving ML-Ready Dataset <br>


### Objective

Save the fully transformed dataset after:
- One-hot encoding
- Feature scaling

This dataset is now fully numeric and ready for machine learning modeling.

The final version is stored as a pickle file (.pkl) for efficient loading in the modeling notebook.

```python
# Save the final ML-ready dataset
df.to_pickle('../Data/processed_data/ml_ready_data.pkl')

# Save feature and target columns separately (cleaner for modeling)
X = df[feature_columns]
y = df[target_columns]

X.to_pickle('../Data/processed_data/X_features.pkl')
y.to_pickle('../Data/processed_data/y_targets.pkl')
```
